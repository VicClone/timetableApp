```python
LECTURE = 'lecture'
PRACTICE = 'practice'
LABORATORY = 'laboratory'
TYPE_LESSON = (LECTURE, PRACTICE, LABORATORY)
DAY_WEEK = ('пн', 'вт', 'ср', 'чт', 'пт')
TIMES = ('8:00', '9:00', '10:00', '11:00', '12:00', '13:00', '14:00', '16:00', '17:00')

d1 = {
    'пн': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'вт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'ср': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'чт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'пт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
}
d2 = {
    'пн': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'вт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'ср': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'чт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'пт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
}
d3 = {
    'пн': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'вт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'ср': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'чт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
    'пт': {
        '8:00': False,
        '9:00': False,
        '10:00': False,
        '11:00': False,
        '12:00': False,
        '13:00': False,
        '14:00': False,
        '16:00': False,
        '17:00': False
    },
}

import random
def _create_timetable():
    random.seed()
    res = {}
    count = random.randrange(1, 5)
    for i in range(0, count):
        day = random.choice(DAY_WEEK)
        count2 = random.randrange(1, 9)
        for i in range(0, count2):
            time = random.choice(TIMES)
            res[day] = {time: False}
    return res



class _Name:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return self.name


class Lesson(_Name):
    def __init__(self, name, types=(), is_type=None):
        super().__init__(name)
        if not types and not is_type:
            raise Exception()
        self._check(types)
        self.types = types
        self.is_type = is_type

    @staticmethod
    def _check(types):
        for _type in types:
            if _type not in TYPE_LESSON:
                raise ValueError('Тип {} не входит в {}'.format(_type, TYPE_LESSON))


class Calendar:
    def __init__(self, data=None):
        """

        Args:
            data (dict): расписание, если нет значеня - занят - {
                "пн": {
                    "9:00": obj - какой-то объект, если занят
                    ...
                }
                ....
            }
        """
        self.data = data

    def __repr__(self):
        tmp = '{}\n'
        tmp2 = '{}:\n'
        day_array = []
        for day, time_table in self.data.items():
            _day = tmp2.format(day)
            day_array.append(_day)
            for time, data in time_table.items():
                if data:
                    day_array.append(
                        tmp.format(
                            '{}: {}'.format(
                                time,
                               ', '.join([str(_data) for _data in (data or [])])
                            )
                        )
                    )

        return ''.join(day_array)
        # return str(id(self))

    def check(self, day, time):
        return not bool(self.data.get(day, {}).get(time))

    def fill(self, day, time, obj=True):
        """
        Заполнить

        Args:
            day (str): день недели
            time (str): время
            obj (object): какая-то информация

        Returns:
            bool: True - успех, иначе False
        """
        if not self.check(day, time):
            return False

        time_table = self.data.get(day, {})
        time_table[time] = obj
        self.data[day] = time_table

        return True

    def get_free_time(self, day):
        return {k for k, v in self.data.get(day, {}).items() if not v}

    def get_not_free_time(self, day):
        return {k for k, v in self.data.get(day, {}).items() if v}


class _NameTimetable(_Name):
    def __init__(self, timetable, *args, **kwargs):
        """

        Args:
            timetable (Calendar): календарь
        """
        super().__init__(*args, **kwargs)
        self.timetable = timetable


class Class(_NameTimetable):
    def __init__(self, count, workload, *args, **kwargs):
        """

        Args:
            name (str): название
            timetable (Calendar): расписание
            count (int): количество
            workload (dict): {день: {предмет: количество}}
        """
        super().__init__(*args, **kwargs)
        self.count = count
        self.workload = workload


class Teacher(_NameTimetable):
    def __init__(self, lesson, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.lesson = lesson

    def check_lesson(self, lesson):
        """

        Args:
            lesson (Lesson): урок

        Returns:
            bool: True если подходит, False иначе

        """
        #TODO можно сдеать сравнение на уровне Lesson
        return self.lesson.name == lesson.name and lesson.is_type in self.lesson.types


class Room(_NameTimetable):
    def __init__(self, types, capacity, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.types = types
        self.capacity = capacity


class Rooms:
    def __init__(self, rooms):
        """

        Args:
            rooms (list of Room): список кабинетов
        """
        self.rooms = rooms

    def find_free(self, day, time, capacity):
        for room in self.rooms:
            if room.capacity >= capacity:
                not_free_time = room.timetable.get_not_free_time(day)
                if time not in not_free_time:
                    return room


class Teachers:
    def __init__(self, teachers):
        """

        Args:
            teachers (list of Teacher): список учителей
        """
        self.teachers = teachers

    def find_free(self, day, time, lesson):
        """
        Ищет свободного учителя в определённый день, время и на определённый предмет
        TODO если 2 урока должны идти подряд, то это можно учесть здесь
        Args:
            day (str): день недели
            time (str): время
            lesson (Lesson): предмет

        Returns:
            Teacher

        """
        for teacher in self.teachers:
            if teacher.check_lesson(lesson):
                free_time = teacher.timetable.get_free_time(day)
                if time in free_time:
                    return teacher


class TimetableCreator:
    def __init__(self, rooms, teachers, classes, timetable):
        """

        Args:
            rooms (Rooms): список аудиторий
            teachers (Teachers): список учителей
            classes (list of Class): список групп
            timetable (Calendar): заполняемы календарь
        """
        self.rooms = rooms
        self.teachers = teachers
        self.classes = classes
        self.timetable = timetable

    def create(self):
        for _class in self.classes:
            for day, lesson2count in _class.workload.items():
                for lesson, count_lesson in lesson2count.items():
                    for _ in range(0, count_lesson):
                        # TODO проверить соотнощение количества свободного времени и нагрузки
                        free_time = _class.timetable.get_free_time(day)
                        for time in free_time:
                            room = self.rooms.find_free(day, time, _class.count)
                            teacher = self.teachers.find_free(day, time, lesson)
                            if room and teacher:
                                self.fill(day, time, teacher, lesson, _class, room)
                                break
            # TODO проверить для класса все ли у него дни распланированны.

    def fill(self, day, time, teacher, lesson, _class, room):
        """

        Args:
            day (str): день
            time (str): время
            teacher (Teacher): преподаватель
            lesson (Lesson): предмет
            _class (Class): класс
            room (Room): кабинет

        Returns:

        """
        self.timetable.fill(day, time, (teacher, lesson, _class, room))
        _class.timetable.fill(day, time, (teacher, lesson, room))
        teacher.timetable.fill(day, time, (_class, lesson, room))
        room.timetable.fill(day, time, (teacher, _class, lesson))


if __name__ == '__main__':
    lesson = Lesson('физика', [LECTURE, PRACTICE])
    lesson2 = Lesson('русский', [LECTURE])
    lesson3 = Lesson('русский', [PRACTICE])
    lesson4 = Lesson('химия', [LECTURE])
    teacher_1 = {
        'пн': {
            '8:00': False,
            '9:00': False,
            '10:00': True,
            '11:00': False,
            '12:00': False,
            '13:00': True,
            '14:00': False,
            '16:00': False,
            '17:00': False
        },
        'вт': {
            '8:00': False,
            '9:00': False,
            '10:00': False,
            '11:00': False,
            '12:00': False,
            '13:00': True,
            '14:00': True,
            '16:00': True,
            '17:00': True
        },
        'ср': {
            '8:00': False,
            '9:00': False,
            '10:00': False,
            '11:00': False,
            '12:00': False,
            '13:00': False,
            '14:00': False,
            '16:00': False,
            '17:00': False
        },
        'чт': {
            '8:00': False,
            '9:00': False,
            '10:00': False,
            '11:00': False,
            '12:00': False,
            '13:00': False,
            '14:00': False,
            '16:00': False,
            '17:00': False
        },
        'пт': {
            '8:00': True,
            '9:00': True,
            '10:00': True,
            '11:00': True,
            '12:00': True,
            '13:00': True,
            '14:00': True,
            '16:00': True,
            '17:00': True
        },
    }
    teachers = []
    teachers.append(Teacher(name='Петров И.Г.', lesson=lesson, timetable=Calendar(teacher_1)))
    teachers.append(Teacher(name='Иванов И.П.', lesson=lesson2, timetable=Calendar(teacher_1)))
    teachers.append(Teacher(name='Иванов И.П.', lesson=lesson3, timetable=Calendar(teacher_1)))
    teachers.append(Teacher(name='Филиппова М.А.', lesson=lesson4, timetable=Calendar(teacher_1)))
    teachers.append(Teacher(name='Лещева И.П.', lesson=lesson2, timetable=Calendar(teacher_1)))
    teachers.append(Teacher(name='Калинин С.М.', lesson=lesson, timetable=Calendar(teacher_1)))
    teachers = Teachers(teachers)
    _class = Class(
        name='5б',
        count=4,
        timetable=Calendar(d1),
        workload={
            'пн': {Lesson(name='физика', is_type=LECTURE): 2,
                   Lesson(name='русский', is_type=LECTURE): 1},
            'вт': {Lesson(name='русский', is_type=LECTURE): 2},
            'ср': {Lesson(name='химия', is_type=LECTURE): 1},
            'чт': {Lesson(name='физика', is_type=LECTURE): 3},
            'пт': {Lesson(name='русский', is_type=LECTURE): 1},
        }
    )
    _class2 = Class(
        name='5a',
        count=40,
        timetable=Calendar(d2),
        workload={
            'пн': {Lesson(name='физика', is_type=LECTURE): 1,
                   Lesson(name='русский', is_type=LECTURE): 1},
            'вт': {Lesson(name='физика', is_type=LECTURE): 2,
                   Lesson(name='физика', is_type=LECTURE): 1},
            'ср': {Lesson(name='физика', is_type=LECTURE): 2,
                   Lesson(name='русский', is_type=LECTURE): 2},
            'чт': {Lesson(name='физика', is_type=LECTURE): 2,
                   Lesson(name='физика', is_type=LECTURE): 3},
            'пт': {Lesson(name='химия', is_type=LECTURE): 1,
                   Lesson(name='русский', is_type=LECTURE): 1},
        }
    )
    _class3 = Class(
        name='5в',
        count=15,
        timetable=Calendar(d3),
        workload={
            'пн': {Lesson(name='химия', is_type=LECTURE): 2,
                   Lesson(name='русский', is_type=LECTURE): 2},
            'вт': {Lesson(name='физика', is_type=LECTURE): 1,
                   Lesson(name='физика', is_type=LECTURE): 1},
            'ср': {Lesson(name='физика', is_type=LECTURE): 2,
                   Lesson(name='русский', is_type=LECTURE): 1},
            'чт': {Lesson(name='физика', is_type=LECTURE): 1,
                   Lesson(name='химия', is_type=LECTURE): 2},
            'пт': {Lesson(name='физика', is_type=LECTURE): 1,
                   Lesson(name='химия', is_type=LECTURE): 2},
        }
    )
    rooms = []
    rooms.append(Room(TYPE_LESSON, 20, name='R205', timetable=Calendar(_create_timetable())))
    rooms.append(Room(TYPE_LESSON, 30, name='40E', timetable=Calendar(_create_timetable())))
    rooms.append(Room(TYPE_LESSON, 40, name='Т405', timetable=Calendar(_create_timetable())))
    rooms = Rooms(rooms)

    classes = [_class, _class2, _class3]

    creator = TimetableCreator(rooms, teachers, classes, Calendar(_create_timetable()))
    creator.create()
    for _class_ in classes:
        print(_class_.name, _class_.timetable)
```
